```python
# -----------------------------
# Day 2: Advanced Insights + Report
# -----------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# 1. Load Data
file_path = r"C:\HR-Attrition-Analytics-project\HR_Attrition_Cleaned.csv"
df = pd.read_csv(file_path)

# 2. Clean Column Names
df.columns = df.columns.str.strip().str.title()

# 3. Open PDF for saving plots
report_path = r"C:\HR-Attrition-Analytics-project\Day2_Insights_Report.pdf"
pp = PdfPages(report_path)

# -----------------------------
# INSIGHTS & PLOTS
# -----------------------------

# 1. Overall Attrition
plt.figure(figsize=(6,5))
sns.countplot(x="Attrition", data=df, hue="Attrition", palette="Set2", legend=False)

plt.title("Overall Attrition Distribution")
plt.show()   # <-- This will display the graph













```

```python
# 2. Attrition by Gender
plt.figure(figsize=(6,5))
sns.countplot(x="Gender", hue="Attrition", data=df, palette="Set2")
plt.title("Attrition by Gender")
plt.show()   # <-- This will display the graph
```

```python
# 3. Attrition by Department
plt.figure(figsize=(7,5))
sns.countplot(x="Department", hue="Attrition", data=df, palette="Set2")
plt.title("Attrition by Department")
plt.xticks(rotation=30)
plt.show()   # <-- This will display the graph
```

```python
# 4. Attrition by Job Role
plt.figure(figsize=(10,6))
sns.countplot(x="Jobrole", hue="Attrition", data=df, palette="Set2")
plt.title("Attrition by Job Role")
plt.xticks(rotation=45)
plt.show()   # <-- This will display the graph
```

```python

# 5. Attrition by Age
plt.figure(figsize=(7,5))
sns.histplot(data=df, x="Age", hue="Attrition", multiple="stack", palette="Set2", bins=20)
plt.title("Attrition by Age Distribution")
plt.show()   # <-- This will display the graph
```

```python
# 6. Monthly Income vs Attrition
plt.figure(figsize=(7,5))
sns.boxplot(x="Attrition", y="Monthlyincome", data=df, hue="Attrition", palette="Set2", legend=False)

plt.title("Attrition vs Monthly Income")
plt.show()   # <-- This will display the graph
```

```python

# 7. Overtime vs Attrition
plt.figure(figsize=(6,5))
sns.countplot(x="Overtime", hue="Attrition", data=df, palette="Set2")
plt.title("Attrition vs Overtime")
plt.show()   # <-- This will display the graph
```

```python
# 8. Years at Company vs Attrition
plt.figure(figsize=(7,5))
sns.histplot(data=df, x="Yearsatcompany", hue="Attrition", multiple="stack", palette="Set2", bins=15)
plt.title("Attrition vs Years at Company")
plt.show()   # <-- This will display the graph
```
